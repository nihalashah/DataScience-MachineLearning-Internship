# Pie Chart

```python
y = np.array([20,35,10,45])

plt.pie(y)
plt.show()
```

```python
#labels
y = np.array([10,20,20,60])
mylabels = ["USA", "India", "Russia", "China"]

plt.pie(y, labels=mylabels,autopct='%1.1f%%')  # autopct for show percentage
plt.show() 
```

```python
#angle set
plt.pie(y, labels = mylabels, startangle = 60)
plt.show() 
```

```python
#explode
myexplode = [0.4, 0.8, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode)
plt.show() 
```

```python
#shoadow
plt.pie(y, labels = mylabels, explode = myexplode, shadow = True)
plt.show()
```

```python
#color
mycolors = ["black", "red", "b", "#4CAF50"]

plt.pie(y, labels = mylabels, colors = mycolors)
plt.show() 
```
